```python
import math
import random

print("Scientific Calculator")

while True:
    print("Options:")
    print("Enter 'add' to add two numbers")
    print("Enter 'sub' to subtract two numbers")
    print("Enter 'mul' to multiply two numbers")
    print("Enter 'div' to divide two numbers")
    print("Enter 'pow' to find power of a number")
    print("Enter 'sqrt' to find square root of a number")
    print("Enter 'sin' to find sine of an angle in radians")
    print("Enter 'cos' to find cosine of an angle in radians")
    print("Enter 'tan' to find tangent of an angle in radians")
    print("Enter 'sin_deg' to find sine of an angle in degrees")
    print("Enter 'cos_deg' to find cosine of an angle in degrees")
    print("Enter 'tan_deg' to find tangent of an angle in degrees")
    print("Enter 'log' to find logarithm of a number with a given base")
    print("Enter 'exp' to find exponential of a number")
    print("Enter 'fact' to find factorial of a number")
    print("Enter 'perm' to find permutations of two numbers")
    print("Enter 'comb' to find combinations of two numbers")
    print("Enter 'abs' to find absolute value of a number")
    print("Enter 'floor' to find floor value of a number")
    print("Enter 'ceil' to find ceiling value of a number")
    print("Enter 'gcd' to find greatest common divisor of two numbers")
    print("Enter 'lcm' to find least common multiple of two numbers")
    print("Enter 'rand' to generate a random number between two numbers")
    print("Enter 'deg2rad' to convert degrees to radians")
    print("Enter 'rad2deg' to convert radians to degrees")
    print("Enter 'quit' to exit the program")
    
    user_input = input(": ")
    
    if user_input == "quit":
        break
        
    elif user_input == "add":
        num1 = float(input("Enter a number: "))
        num2 = float(input("Enter a number: "))
        print(num1 + num2)
                           
                           
    elif user_input == "sub":
        num1 = float(input("Enter a number: "))
        num2 = float(input("Enter another number: "))
        print(num1 - num2)
        
    elif user_input == "mul":
        num1 = float(input("Enter a number: "))
        num2 = float(input("Enter another number: "))
        print(num1 * num2)
        
    elif user_input == "div":
        num1 = float(input("Enter a number: "))
        num2 = float(input("Enter another number: "))
        print(num1 / num2)
        
    elif user_input == "pow":
        num1 = float(input("Enter a number: "))
        num2 = float(input("Enter power: "))
        print(math.pow(num1, num2))
        
    elif user_input == "sqrt":
        num1 = float(input("Enter a number: "))
        print(math.sqrt(num1))
        
    elif user_input == "sin":
        num1 = float(input("Enter an angle in radians: "))
        print(math.sin(num1))
        
    elif user_input == "cos":
        num1 = float(input("Enter an angle in radians: "))
        print(math.cos(num1))
        
    elif user_input == "tan":
        num1 = float(input("Enter an angle in radians: "))
        print(math.tan(num1))
        
    elif user_input == "log":
        num1 = float(input("Enter a number: "))
        base = float(input("Enter the base of the logarithm: "))
        print(math.log(num1, base))
        
    elif user_input == "exp":
        num1 = float(input("Enter a number: "))
        print(math.exp(num1))
        
    elif user_input == "fact":
        num1 = int(input("Enter a number: "))
        print(math.factorial(num1))
    
    elif user_input == "abs":
        num1 = float(input("Enter a number: "))
        print(abs(num1))
    
    elif user_input == "floor":
        num1 = float(input("Enter a number: "))
        print(math.floor(num1))
    
    elif user_input == "ceil":
        num1 = float(input("Enter a number: "))
        print(math.ceil(num1))
    
    elif user_input == "gcd":
        num1 = int(input("Enter a number: "))
        num2 = int(input("Enter another number: "))
        print(math.gcd(num1, num2))
    
    elif user_input == "lcm":
        num1 = int(input("Enter a number: "))
        num2 = int(input("Enter another number: "))
        lcm = (num1 * num2) // math.gcd(num1, num2)
        print(lcm)
   
    elif user_input == "deg2rad":
        num1 = float(input("Enter an angle in degrees: "))
        rad = math.radians(num1)
        print(rad)
    
    elif user_input == "rad2deg":
        num1 = float(input("Enter an angle in radians: "))
        deg = math.degrees(num1)
        print(deg)
                                   
```
